We define the class `Normaliser` with the functions `normalise1D`, to work with normalised data, and `unnormalise1D` to see the predictions in real scale later. Having a class for this is useful for this problem, as `x_normaliser` and `y_normaliser` save the values from the original vectors x and y respectively, so the transformations can always be made w/r to them. 

```python
class Normaliser:
    def __init__(self,z , new_min = 0, new_max = 1):
        self.new_min = 0
        self.new_max = 1
        self.z_min = np.min(z)
        self.z_max = np.max(z)
    
    def normalise1D(self,z): 
        z_normalised = self.new_min + (z - self.z_min)*(self.new_max - self.new_min)/(self.z_max - self.z_min) 
        return z_normalised
    
    def unnormalise1D(self,z_norm):
        z = (z_norm - self.new_min)*(self.z_max - self.z_min)/(self.new_max - self.new_min) + self.z_min
        return z

#Save values from original vectors x and y
x_normaliser = Normaliser(x)
y_normaliser = Normaliser(y)
```

```python
x_norm = x_normaliser.normalise1D(x)
y_norm = y_normaliser.normalise1D(y)
```

## 2.1 Bayesian Linear Regression predictions

We define BLR functions to make the tuning process of the parameters alpha, beta, and p easier.

```python
def generate_feature_mat(x_in, p):
    theta_x = x_in**np.arange(p+1)
    return theta_x

def S_N(alpha, beta, xtrain, p):
    return np.linalg.pinv(alpha*np.eye((p+1)) + beta*generate_feature_mat(xtrain,p).T.dot(generate_feature_mat(xtrain,p)))

def m_N(alpha, beta, xtrain, p, ytrain):
    return beta*S_N(alpha, beta, xtrain, p) @ generate_feature_mat(xtrain,p).T @ ytrain

def mu_q(alpha, beta, xtrain, p, ytrain,xpred):
    mean = m_N(alpha, beta, xtrain, p, ytrain).T.dot(generate_feature_mat(xpred, p).T).ravel()
    return mean

def sigma_q(alpha, beta, xtrain, p, ytrain, xpred):    
    stdev = np.sqrt(np.diag((generate_feature_mat(xpred, p).dot(S_N(alpha, beta, xtrain, p)).dot(generate_feature_mat(xpred, p).T))) + 1.0 / beta)
    return stdev


#def generate_feature_mat_fourier(x_in,r,w):
#    matrix = np.empty((0, 4), int)
#    for i in range(1,r+1,2):
#        a = np.sin(i*w*x_in)/(np.sqrt(2*np.pi/w)/2)
#        matrix=np.append(matrix,a)
#    for i in range(0,r+1,2):
#        b = np.cos(i*w*x_in)/(np.sqrt(2*np.pi/w)/2)
#        matrix=np.append(matrix,b)
#    theta_x = np.reshape(matrix,(r+1,len(x_in))).T
#    return theta_x
```

First, we'll run the prediction for the whole space so we can visualize the fit and manually tune the model (for parameters $\alpha, \beta$ and p).

```python
alpha = 5
beta = 30
p = 2

x_q = np.linspace(-0.1, 1.1, 200)[:,np.newaxis] 
mu_pred = mu_q(alpha, beta, x_norm, p, y_norm, x_q)
variance_pred = sigma_q(alpha, beta, x_norm, p, y_norm, x_q)**2
sigma_pred = np.sqrt(variance_pred)

pl.fill_between(x_q.ravel(), mu_pred-2*sigma_pred, mu_pred+2*sigma_pred, \
                facecolor='green', interpolate=True, alpha=0.5, label='2*sigma of prediction')
pl.scatter(x_norm, y_norm, label='ground truth')
pl.plot(x_q, mu_pred, c='red', label='mu of prediction')
pl.legend(loc='lower right')
pl.title('Bayesian linear regression ')
```

After manually experimenting with values, the chosen parameters are $\alpha = 5$, $\beta = 30$, and $p = 2$. There was a very simple and specific criteria for each one of them:

$\alpha$: Since our data is strong (not sparse), and we don't have confidence on our prior knowledge for this problem, we selected a value for $\alpha$  that is small enough so that the model is mainly learning from the data (i.e. $m_N$ is similar to the likelihood), but big enough so that our prior still has some influence on it and it isn't improper.

$\beta$: This parameter is inversely related to the variance, and allows us to determine how noisy we think the target distribution is (so for bigger $\beta$, less noisy). In this case, the data has a very reliable pattern, and the desired predictions are either close in the future or inside the range. This motivated a high $\beta$ that spans a standard deviation such that it just barely fits every point of the known data set within a maximum of two standard deviations from the predicted expected value.  

$p$: determines the polynomial degree of the basis functions (w/r to x). A second degree polynomial for this case allows the model to capture the macro trend without overfitting. 


We now normalise the data points in order to make the predictions

## 2.2 Gaussian Processes mean sea level predictions

The squared exponential kernel was chosen to model this distribution, because the goal is not to predict every single wiggle (which would result in overfitting), but to capture the general trend of the distribution and set a variance that accounts for the wiggles. The general trend is a very smooth distribution, and has no discontinuities, so the squared exponential kernel is very suitable

```python
class GP:
    def __init__(self,cov_function_name, x_train, y_train, hyper_params, sigma_n):
        self.cov_function_name = cov_function_name
        self.x_train = x_train
        self.y_train = y_train
        self.hyper_params = hyper_params #params[0] is signal variance, params[1] is lengthscale
        self.K = self.cov_func(self.x_train,self.x_train)
        self.sigma_n = sigma_n
        self.K_inv = np.linalg.pinv(self.K+np.eye(self.K.shape[0])*(self.sigma_n**2))
        
    def predict(self,x_star):
        k_star = self.cov_func(self.x_train,x_star)
        f_mean = np.dot(np.dot(np.transpose(k_star),self.K_inv),self.y_train)
        f_std = self.cov_func(x_star, x_star) - np.dot(np.dot(np.transpose(k_star),self.K_inv),k_star)+np.eye(x_star.shape[0])*(self.sigma_n**2)
        f_std = np.sqrt(np.diag(f_std))[:,np.newaxis]
        return [f_mean,f_std]

    def cov_func(self,x_1,x_2):
        if self.cov_function_name == "Squared Exponential":
            x_1_sq = np.sum(np.square(x_1),1)
            x_2_sq = np.sum(np.square(x_2),1)
            d= -2*np.dot(x_1, x_2.T) + (x_1_sq[:,None] + x_2_sq[None,:])
            d= np.sqrt(np.clip(d,0,np.inf))/self.hyper_params[1]
            K = self.hyper_params[0] * np.exp(-0.5*d**2)  
        return K
    
    def set_hyper_params(self, hyper_params, sigma_n):
        self.hyper_params = hyper_params
        self.K = self.cov_func(self.x_train,self.x_train)
        self.sigma_n = sigma_n
        self.K_inv = np.linalg.pinv(self.K+np.eye(self.K.shape[0])*(self.sigma_n**2))
        
    def log_marginal_likelihood(self):
        lml1 = -.5*np.dot(np.dot(np.transpose(self.y_train),self.K_inv),self.y_train)
        lml2 = -.5*np.log(np.linalg.det(self.K+np.eye(self.K.shape[0])*(self.sigma_n**2)))
        lml3 = -.5*self.x_train.shape[0]*np.log(2*np.pi)
        return lml1+lml2+lml3
```

First, we will tune the model to select a lengthscale and a $\sigma_n$. The lengthscale in this case plays a very similar role to the p in the BLR, which determined the grade of the polynomial of the basis functions, but in the opposite direction (a smaller lengthscale will make the model more adaptable to small wiggles in the data, as opposed to p which is more adaptable when it's bigger). $\sigma_n$ plays a similar role to the inverse of $\beta$ in the BLR, so we will use the same criteria as in the last section to tune those two parameters.

The case of the signal variance is different, because there is no analogous parameter in the BLR. The signal variance determines how free the model is to "chase outliers", ie, grow its variance in certain spots where the data is further away from the mean. For this particular distribution, the signal variance will be set so it's effect is low compared to $\sigma_n$, as the latter will be set in order to "capture" the data within two standard deviations of the mean, which only follows the bigger trend. This means a high value of signal variance for this model would be redundant, because there won't be any outliers to "chase".

```python
signal_variance = 10
lengthscale = 0.9
sigma_n = 0.15

x_star = np.linspace(0,1.1,100)[:,np.newaxis] 
#x_star = x_prednorm

gp = GP("Squared Exponential", x_norm, y_norm, [signal_variance,lengthscale], sigma_n)
[f_mean_2,f_std_2] = gp.predict(x_star)
lml = gp.log_marginal_likelihood()
print("Log Marginal Likelihood: "+str(lml))

pl.scatter(x_norm, y_norm)
pl.fill_between(x_star.ravel(), (f_mean_2-2*f_std_2).ravel(), (f_mean_2+2*f_std_2).ravel(), \
                facecolor='blue', interpolate=True, alpha=0.1, label='2*sigma of prediction')
pl.plot(x_star, f_mean_2)
pl.title('Mean sea level over time', fontsize=14)
pl.xlabel('x [normalised]')
pl.ylabel('y [normalised and zero mean]')
pl.grid()
```

```python
gp.set_hyper_params([1,0.8],sigma_n)
[f_mean_1,f_std_1] = gp.predict(x_star)

gp.set_hyper_params([1,0.7],sigma_n)
[f_mean_2,f_std_2] = gp.predict(x_star)

gp.set_hyper_params([1,0.9],sigma_n)
[f_mean_3,f_std_3] = gp.predict(x_star)


fig, ax = pl.subplots(1,3,figsize=(5*4,4))

ax[0].scatter(x_norm, y_norm)
ax[0].fill_between(x_star.ravel(), (f_mean_1-2*f_std_1).ravel(), (f_mean_1+2*f_std_1).ravel(), \
                facecolor='blue', interpolate=True, alpha=0.1, label='2*sigma of prediction')
ax[0].plot(x_star, f_mean_1)
ax[0].set_title('lengthscale = 0.8')
ax[0].grid()

ax[1].scatter(x_norm, y_norm)
ax[1].fill_between(x_star.ravel(), (f_mean_2-2*f_std_2).ravel(), (f_mean_2+2*f_std_2).ravel(), \
            facecolor='blue', interpolate=True, alpha=0.1, label='2*sigma of prediction')
ax[1].plot(x_star, f_mean_2)
ax[1].set_title('lengthscale = 0.7')
ax[1].grid()

ax[2].scatter(x_norm, y_norm)
ax[2].fill_between(x_star.ravel(), (f_mean_3-2*f_std_3).ravel(), (f_mean_3+2*f_std_3).ravel(), \
                facecolor='blue', interpolate=True, alpha=0.1, label='2*sigma of prediction')
ax[2].plot(x_star, f_mean_3)
ax[2].set_title('lengthscale = 0.9')
ax[2].grid()
```

# 3. Suitability of each method

For this particular data set, both methods performed similarly in terms of how they could adapt to fit the data, and the resulting predictions. In fact, with the right choice of parameters, we could have obtained equal predictions. 

The only big difference between the two, in this regard, is the signal variance parameter in the Gaussian Process, which allows us to determine how the model should adapt its predicted variance for outliers. We modeled the data to predict the macro trend this time, but if we want to take a different approach, and model every wiggle of the data, which had many, this parameter would play a key role. 

We only had access to this parameter because we chose the squared exponential kernel. There is no basis function for a BLR that can resemble the squared exponential kernel, as this kernel would be like using an infinite-dimensional basis vector, which would make the computational times infinite. Thanks to the kernel trick, we can do this in a finite amount of time when working with GP's. 

Despite the latter, GP's are generally more computationally expensive than BLR's, and it was the case in this implementation. This would only be false if we had set a big enough p (in the thousands) for the basis functions in the BLR. It is also worth mentioning that the computational time difference in this case wasn't noticeable because N was very small, but as N gets larger, GP's performance scale very poorly, in terms of computational time and memory.

The models are also the same in the sense that they are both Bayesian. They both consider priors, and they both predict accounting for uncertainty.
